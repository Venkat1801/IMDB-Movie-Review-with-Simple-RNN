```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential

# Example data
sent = [
    'the glass of milk',
    'the glass of juice',
    'iam a good boy',
    'iam a good developer',
    'understand the meaning of words',
    'your videos are good'
]

# 1) Set vocabulary size
voc_size = 10000

# 2) One-hot encode sentences
one_hot_rep = [one_hot(words, voc_size) for words in sent]

# 3) Set fixed sentence length BEFORE creating model
sent_length = 8

# 4) Pad sequences
embedded_docs = pad_sequences(one_hot_rep, padding='pre', maxlen=sent_length)
print("Padded Sequences:\n", embedded_docs)

# 5) Set embedding dimension BEFORE creating model
dim = 10

# 6) CREATE MODEL NOW (after variables exist)
model = Sequential()
model.add(Embedding(voc_size, dim, input_length=sent_length))
model.compile(optimizer='adam', loss='mse')

# 7) Check model
model.summary()
```

```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np

sent = [
    'the glass of milk',
    'the glass of juice',
    'iam a good boy',
    'iam a good developer',
    'understand the meaning of words',
    'your videos are good'
]

voc_size = 10000
one_hot_rep = [one_hot(words, voc_size) for words in sent]

sent_length = 8
embedded_docs = pad_sequences(one_hot_rep, padding='pre', maxlen=sent_length)
print(embedded_docs)

dim = 10

model = Sequential()
model.add(Embedding(voc_size, dim, input_length=sent_length))
model.compile(optimizer='adam', loss='mse')

# build model before summary
model.build(input_shape=(None, sent_length))
model.summary()
```

```python
model.predict(embedded_docs)
```
